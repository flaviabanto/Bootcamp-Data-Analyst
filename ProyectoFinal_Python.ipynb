##Partidos del mundial con más goles

```python
#Obteniendo las columnas mas relevantes
df_matches.head()
df_matches_relevant = df_matches.get(['Year', 'Home Team Name', 'Away Team Name', 'Home Team Goals', 'Away Team Goals'])
df_matches_relevant.head()
```

```python
#Agregamos una columna con los goles totales del partido
df_matches_relevant['Total Goals'] = df_matches_relevant['Home Team Goals'] + df_matches_relevant['Away Team Goals']
df_matches_relevant.head()
```

```python
#Ordenamos d emanera descendente y mostramos los 20 partidos con mas goles
df_matches_relevant.sort_values(by = 'Total Goals', ascending = False).head(20)
```

```python
#Los partidos con mas goles, del año 2000 en adelante
df_matches_relevant.sort_values(by = 'Total Goals', ascending = False)\
                  .loc[df_matches_relevant['Year']>2000].head(20)
```

##Goleadores históricos de los mundiales

```python
df_players.head()
```

```python
#Eliminar valores nulos del dataframe usamos dropna
df_players_events = df_players.dropna(subset = ['Event'])
df_players_events.head(10)
```

```python
#La cantidad de goles por jugador se ve en la cantidad de G que aparecen en la columna event, así que usaremos una funcion
def cantidad_goles(cadena):
  return cadena.count('G')

cantidad_goles("G43' G87'") 
```

```python
#Ahora creamos una nueva columna con la cantidad de goles por jugador
df_players_events['NumGoals'] = df_players_events['Event'].apply(cantidad_goles)
df_players_events.head(100)
```

```python
#ahora nos quedamos con las columnas relevantes
df_players_events.head()
df_players_events.get(['Player Name', 'NumGoals'])
```

```python
#vamos a agrupar por el nombre de cada jugador, porque se ve por la cantidad de goles en todos los partidos, no solo en uno
df_players_events.head()
df_players_events.get(['Player Name', 'NumGoals'])\
                 .groupby('Player Name')\
                 .sum('NumGoals')\
                 .sort_values(by = 'NumGoals', ascending = False).head(10)
```
